```python
import pandas as pd 
from funpymodeling import freq_tbl, status
```

```python
game_path = 'C:/Users/Administrador/Desktop/Game/Data/Video_Games_Sales_as_at_22_Dec_2016.csv'
data = pd.read_csv(game_path, sep= ',')
```

```python
status(data)
```

## Descarto columnas irrelevantes

```python
data.drop(columns=['User_Count', 'Critic_Count', 'Developer'], inplace=True)
```

```python
status(data)
```

## Elimino filas donde faltan los valores mas importantes

```python
data = data.dropna(subset=['Rating', 'Critic_Score', 'User_Score'])
```

#### Convierto tipo de datos

La columna de User_Score debe ser numerica

```python
data['User_Score'] = pd.to_numeric(data['User_Score'], errors='coerce')
```

```python
status(data)
```

### Variables categoricas

```python
data = pd.get_dummies(data, columns= ['Platform', 'Genre', 'Rating'], drop_first=True)
```

```python
status(data)
```

```python
data
```

Convierto todos los bools en 0 y 1. 

```python
bool_cols = data.select_dtypes(include=['bool']).columns
data[bool_cols] = data[bool_cols].astype(int)
```

```python
data
```

# Clasificación 

### Separacion X de Y

```python
data_x = data.drop(columns=["Name", "Publisher", "Global_Sales"], axis=1)
data_y = data["Global_Sales"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
```

## Creacion del modelo predictivo

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=1000, random_state=99)
```

```python
rf.fit(x_train, y_train)
```

## Evaluacion del modelo

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = rf.predict(x_test)

print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
