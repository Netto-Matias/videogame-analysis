#### Ventas globales vs Ventas regionales

```python
correlacion = data[['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlacion entre ventas por regiones y ventas globales')
plt.show()
```
